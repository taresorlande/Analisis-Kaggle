```python
df = pd.read_csv(os.path.join(dirname, filenames[0]))

freedom = df["Freedom"].tolist()
happiness = df["Happiness Score"].tolist()
family = df["Family"].tolist()

print(freedom)
print()
print(happiness)
print()
print(family)

# print(math.floor(min(happiness)))
# print(math.ceil(max(happiness)))
```

```python
# https://stackoverflow.com/questions/6148207/linear-regression-with-matplotlib-numpy
import numpy as np
import matplotlib.pyplot as matplotlib

x = freedom
y = happiness

coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef)
# poly1d_fn is now a function which takes in x and returns an estimate for y

matplotlib.plot(x,y, 'yo', x, poly1d_fn(x), '--k')
matplotlib.title('Linear Regression')
matplotlib.xlim(min(freedom), max(freedom))
matplotlib.ylim(min(happiness), max(happiness))
```

```python
import numpy as np
import pandas as pd
import matplotlib.cm as cm
import matplotlib.pyplot as plt

family = np.array(family)
happiness = np.array(happiness)

def scale(a):
    return (a-np.min(a))/(np.max(a) - np.min(a))
scaled = scale(happiness*family)

for x, y, z in zip(family, happiness, scaled):
    plt.scatter(x, y, color = cm.rainbow(z))
    
plt.title('Clustering')
```

```python
#converte a tabela dinâmica apresentada por um dicionário Python em uma string formatada em JSON
webdatarocks_json_object = json.dumps(pivot_table_configuration)
```

```python
#função que renderiza a tabela dinâmica e os gráficos na célula do notebook
def render_table_and_charts(webdatarocks_json_object, draw_charts=''''''):
    
    code = '''
      <link href="https://cdn.webdatarocks.com/latest/webdatarocks.min.css" rel="stylesheet"/>
      <script src="https://cdn.webdatarocks.com/latest/webdatarocks.js"></script>
      <script src="https://cdn.webdatarocks.com/latest/webdatarocks.googlecharts.js"></script>
      <script src="https://www.gstatic.com/charts/loader.js"></script>
      <h1 style="margin-bottom: 20px;">World Happiness Report Results</h1>
      <div id="pivot-container"></div>
      <div style="margin-top: 20px;" id="googlechart-container"></div>
      <script>
      var pivot = new WebDataRocks({0});
      
      {1}
      </script>
      '''.format(webdatarocks_json_object, draw_charts)
    return HTML(code)
```

```python
#Aqui, especificamos o layout HTML que contém os scripts do Google Charts e WebDataRocks, 
#estilos CSS e contêineres onde a tabela dinâmica e as instâncias do gráfico serão renderizadas. 
#Todo o layout está entre aspas triplas.
draw_charts = ''' 
var pivotTableReportComplete = false;
var googleChartsLoaded = false;
pivot.on("reportcomplete", () => {
    pivot.off("reportcomplete");
    pivotTableReportComplete = true;
    createGoogleChart();
});
google.charts.load('current', {
    'packages': ['geochart'],
    'mapsApiKey': 'AIzaSyD-9tSrke72PouQMnMX-a7eZSW0jkFMBWY'
});
google.charts.setOnLoadCallback(onGoogleChartsLoaded);
function onGoogleChartsLoaded() {
    googleChartsLoaded = true;
    if (pivotTableReportComplete) {
        createGoogleChart();
    }
}
function createGoogleChart() {
    if (googleChartsLoaded) {
        pivot.googlecharts.getData({
                type: "bar"
            },
            drawChart,
            drawChart
        );
    }
}
function drawChart(_data) {
    var data = google.visualization.arrayToDataTable(_data.data);
    var options = {
        colorAxis: {
            colors: ['#fff7bc', '#fec44f', '#d95f0e']
        },
        backgroundColor: '#b3e5fc',
        datalessRegionColor: '#ffffff',
        defaultColor: '#f5f5f5'
    };
    var chart = new google.visualization.GeoChart(document.getElementById('googlechart-container'));
    chart.draw(data, options);
}
'''
```

```python
#função de renderização e ver o painel resultante
render_table_and_charts(webdatarocks_json_object, draw_charts)
```
